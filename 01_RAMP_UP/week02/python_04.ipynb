````python
def valid_input(choice, valid_options=None):
    if choice in valid_options:
        return True
    else:
        return False
````

````python
# Realizamos el proceso completo utilizando la forma recursiva
def prompt_user(message, valid_options=None):
    choice = input(message)
    if valid_options:
        if valid_input(choice, valid_options):
            return choice
        else:
            print('input no valido, prueba nuevamente')
            return prompt_user(message, valid_options)
    else:
        return choice
````

````python
## Ejemplo de Evaluación multiple
## Double list comprehensions
countries = ["Brasil", "Italy", "Morocco", "Spain", "France"]
friends = ["Alice", "Bob", "Eve", "John"]
````

````python
%%timeit
pairs = []
for c in countries:
    for f in friends:
        pairs.append(f"{f} visit {c}")
````

````python
%%timeit
[f"{f} visit {c}" for f in friends for c in countries ]
````
