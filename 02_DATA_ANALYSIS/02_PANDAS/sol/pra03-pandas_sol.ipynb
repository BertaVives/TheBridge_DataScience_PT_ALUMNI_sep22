### Ejercicio 5

Añadid una nueva columna al _dataframe_ con un valor booleano indicando una predicción básica de si en una cierta batalla el bando atacante será vencedor. Definiremos el valor como _True_ sí el bando atacante es mayor al defensor, y como _False_ en caso contrario. Definid como _NaN_ aquellos casos en los que no haya número de tropas en ningún bando. ¿En que casos, identificados por el número de batalla (*battle\_number*), nuestra predicción coincide con el resultado real?

```python
# Respuesta
# Creamos una nueva columna
df['nombre_col'] = 0
```

```python
# Para poder asignar un valor en función de una condición (columna/fila u otra)
lambda <arg> : <return value if condition is True> if <condition> else <Return value if condition is False>
```

```python
df_prediccion = df.assign(prediccion=lambda x: x.attacker_size > x.defender_size)
df_prediccion.loc[(df_prediccion['attacker_size'] == 0) & (df_prediccion['defender_size'] == 0), 'prediccion'] = 'NaN'

df_prediccion
```

```python
counter = 0
for row in range(len(df_prediccion['battle_number'])):
  # Establecemos la primera condición, para cada fila, si el resultado de la batalla fue "win" y el resultado de la 
    # predicción fue "True", mostramos por pantalla el "battle_number" de la fila.
    if df_prediccion.attacker_outcome[row] == "win" and df_prediccion.prediccion[row] == "True":
        print(f"Nuestra predicción se cumple para la batalla número: {df_prediccion['battle_number'][row]}")
        counter += 1
    # Establecemos la segunda condición, para cada fila, si el resultado de la batalla fue "loss" y el resultado de la 
    # predicción fue "False", mostramos por pantalla el "battle_number" de la fila.
    elif df_prediccion.attacker_outcome[row] == "loss" and df_prediccion.prediccion[row] == "False":
        print(f"Nuestra predicción se cumple para la batalla número: {df_prediccion['battle_number'][row]}")
        counter += 1

# Mostramos cuántas veces hemos predicho de forma correcta el resultado de las batallas.
print("###"*10)
print(f"Nuestra predicción se cumple {counter} veces")
```

```python
import numpy as np

# creamos las condiciones y las alternativas
conditions = [
              (df.attacker_size > df.defender_size),
              (df.attacker_size < df.defender_size),
              (pd.isna(df.attacker_size) & pd.notna(df.defender_size)),
              (pd.notna(df.attacker_size) & pd.isna(df.defender_size))
]

choices = ["True", "False", "False", "True"]
# Creamos la nueva columna con la función `select` de numpy y las condiciones/alternativas anteriormente creadas. Rellenamos
# la columna por defecto con "NaN", que se rellenará en los casos en los que ningún bando tenga tropas, para el resto de
# posibilidades se rellenarán valores "True" o "False" según las condiciones establecidas.
df['winner_prediction'] = np.select(conditions, choices, default = "NaN")
```

```python
df
```

```python
# Para identificar si nuestra predicción coincide con el resultado real, iteramos sobre el total de filas.
# Creamos un contador para contabilizar el número de predicciones correctas.

counter = 0
for row in range(len(df['battle_number'])):
    # Establecemos la primera condición, para cada fila, si el resultado de la batalla fue "win" y el resultado de la 
    # predicción fue "True", mostramos por pantalla el "battle_number" de la fila.
    if df.attacker_outcome[row] == "win" and df.winner_prediction[row] == "True":
        print(f"Nuestra predicción se cumple para la batalla número: {df['battle_number'][row]}")
        counter += 1
    # Establecemos la segunda condición, para cada fila, si el resultado de la batalla fue "loss" y el resultado de la 
    # predicción fue "False", mostramos por pantalla el "battle_number" de la fila.
    elif df.attacker_outcome[row] == "loss" and df.winner_prediction[row] == "False":
        print(f"Nuestra predicción se cumple para la batalla número: {df['battle_number'][row]}")
        counter += 1

# Mostramos cuántas veces hemos predicho de forma correcta el resultado de las batallas.
print("###"*10)
print(f"Nuestra predicción se cumple {counter} veces")
```
