````python
# Al tratar los datos podemos eliminar valores nulos o NA o NaN
# inplace=True aplica los cambios directamente al dataframe

df_marvel.dropna(axis=1, how='any', inplace=False) 
# how='any' cualquier elemento nulo será eliminada la fila o columna
# how='all' toda la fila o columna será eliminada cuando todos los valores son nulos
````

````python
# Podemos realizar agrupaciones

grouped = df_marvel.groupby(['ALIGN', 'ID'])
for n, g in grouped:
    print(n, len(g))
````

````python
import numpy as np
grouped.aggregate(np.mean)
````

````python
type(grouped.aggregate(np.mean))
````

````python
# Guardamos el formato aggregate a un csv con MultiIndex
grouped.aggregate(np.max).to_csv('data/fichero_marvel_out.csv', encoding='utf-8')
````

## Interacción con base de datos

````python
cols = df_marvel.columns
cols = ", ".join(cols)
````

````python
cols
````

````python
# importamos la librería de Python para crear una base de datos en memoria
import sqlite3

# Creamos un enlace de conexión con la base de datos
conn = sqlite3.connect('marvel.db')

# Con la conexión a través del cursor
cur = conn.cursor()

# Creamos una tabla utilizando los nombres del dataframe con la siguiente Query
sql_create_tabla = "CREATE TABLE marvel_chars ({})".format(cols)
print(sql_create_tabla)

# Lanzamos la sentencia SQL de creación de tabla
cur.execute(sql_create_tabla)

# Guardar cambios
conn.commit()
````

````python
# Ejecutamos una query con SELECT para comprobar la tabla
sql_select_tabla = "SELECT * FROM marvel_chars"
cur.execute(sql_select_tabla)

# Recuperamos la información del resultado de la query o sentencia
results = cur.fetchall()

# Mostramos los resultados
print(results)
````

````python
# Añadimos datos a la tabla
for index, row in df_marvel.iterrows():
    # Por cada fila del DataFrame añadimos la fila a la tabla
    cur.execute("INSERT INTO marvel_chars VALUES "
        "(?,?,?,?,?,?,?,?,?,?,?,?,?)" , row
        )

# Hacemos commit
conn.commit()
````

````python
# Ejecutamos una query con SELECT para comprobar la tabla
sql_select_tabla = "SELECT * FROM marvel_chars"
cur.execute(sql_select_tabla)

# Recuperamos la información del resultado de la query o sentencia
results = cur.fetchall()

# Mostramos los resultados
print(results)
````

````python
# Volvemos a cargar los datos de una tabla DB a DataFrame
pd.read_sql_query(sql_select_tabla, conn)
````

````python
# Cerraremos la conexión de la base de datos

conn.close()
````
